```python
# LET'S FIND OUTLIERS IN THE DATASET SINCE WE HAVE SOME CATEGORICAL COLUMN SO WE HAVE TO SEPRATE IT FROM NUMERICAL COLUMN 
numerical_col=["AveragePrice","Total Volume","4046","4225","4770","Total Bags","Small Bags","Large Bags","XLarge Bags","year"]
plt.figure(figsize=(20,15))
plotnumber=1
for feature in numerical_col:
    if plotnumber<=12:
        ax=plt.subplot(4,3,plotnumber)
        sns.boxplot(avacado[feature])
        plt.xlabel(feature,fontsize=15)
    plotnumber+=1
plt.show()
```

AS WE CAN SEE THAT ONLY YEAR COLUMN HAVE NO OUTLIERS AND REST ONE CONTAIN OUTLIERS SO WE HAVE TO REMOVE THE OUTLIERS SO WILL USE Z- SCORE METHOD 

# Z - SCORE

```python
# LET'S USE Z- SCORE METHOD TO REMOVE THE OUTLIERS
feature_with_outliers= avacado[["AveragePrice","Total Volume","4046","4225","4770","Total Bags","Small Bags","Large Bags","XLarge Bags"]]
```

```python
from scipy.stats import zscore
z=np.abs(zscore(feature_with_outliers))
```

```python
z
```

as you can see that we use Z-SCore and reduce the dataset 

```python
# We have assumed the Z-score threshold < 3 will have no outliers 

avacado_new=avacado[(z<3).all(axis=1)]
```

# LINEAR REGRESSOR

```python
def select_best_random_state(x, y):
    best_random_state =0
    best_r2_score =0

    for random_state in range(1,300):
        X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=random_state)
        LR = LinearRegression()
        LR.fit(X_train, y_train)
        y_pred = LR.predict(X_test)
        r2 = r2_score(y_test, y_pred)
        if r2 > best_r2_score:
            best_r2_score = r2
            best_random_state = random_state

    return best_random_state, best_r2_score


best_random_state, best_r2_score= select_best_random_state(x, y)
print("Best Random State  :", best_random_state)
print("Best r2_score:- ",best_r2_score*100)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=196)
LR=LinearRegression()
LR.fit(X_train, y_train)
y_pred = LR.predict(X_test)
y_pred_train=LR.predict(X_train)
print("r2_score om training dataset",r2_score(y_train,y_pred_train)*100)
print("mean_squared_error",mean_squared_error(y_test,y_pred))
print("men absolute eroor",mean_absolute_error(y_test,y_pred))
print("mean_squared_error",np.sqrt(mean_squared_error(y_test,y_pred)))
```

# RANDOM FOREST REGRESSOR 

```python
def select_best_random_state(x, y):
    best_random_state =0
    best_r2_score =0

    for random_state in range(1,200):
        X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=random_state)
        RFR=RandomForestRegressor()
        RFR.fit(X_train, y_train)
        y_pred = RFR.predict(X_test)
        r2 = r2_score(y_test, y_pred)
        if r2 > best_r2_score:
            best_r2_score = r2
            best_random_state = random_state

    return best_random_state, best_r2_score


best_random_state, best_r2_score= select_best_random_state(x, y)
print("Best Random State  :", best_random_state)
print("Best r2_score:- ",best_r2_score*100)
```
